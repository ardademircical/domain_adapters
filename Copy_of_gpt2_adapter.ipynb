```python
block_size = 128

def group_texts(examples):
    # Concatenate all texts.
    concatenated_examples = {k: sum(examples[k], []) for k in examples.keys()}
    total_length = len(concatenated_examples[list(examples.keys())[0]])
    # We drop the small remainder, we could add padding if the model supported it instead of this drop, you can
    # customize this part to your needs.
    if total_length >= block_size:
        total_length = (total_length // block_size) * block_size
    # Split by chunks of block_size.
    result = {
        k: [t[i : i + block_size] for i in range(0, total_length, block_size)]
        for k, t in concatenated_examples.items()
    }
    result["labels"] = result["input_ids"].copy()
    return result
```

```python
from transformers import AutoModelForCausalLM, TrainingArguments, Trainer
from torch import nn

class DomainController(nn.Module):
  def __init__(self, current_batch_domain=None):
        super().__init__()
        self.current_batch_domain = current_batch_domain

class DomainAdapter(nn.Module):
    """
    The adapters first project the original
    d-dimensional features into a smaller dimension, m, apply
    a nonlinearity, then project back to d dimensions.
    """
    def __init__(self, size = 128, model_dim = 768):
        super().__init__()
        self.debugger = 1
        self.model_dim = model_dim
        self.size = size
        self.adapter_block = nn.Sequential(
            nn.Linear(model_dim, size),
            nn.ReLU(),
            nn.Linear(size, model_dim)
        )

    def forward(self, x):

        ff_out = self.adapter_block(x)
        adapter_out = ff_out + x
        return adapter_out


class Adapter(nn.Module):
    """
    The adapters first project the original
    d-dimensional features into a smaller dimension, m, apply
    a nonlinearity, then project back to d dimensions.
    """
    def __init__(self, size = 128, model_dim = 768):
        super().__init__()
        self.model_dim = model_dim
        self.size = size
        self.adapter_block = nn.Sequential(
            nn.Linear(model_dim, size),
            nn.ReLU(),
            nn.Linear(size, model_dim)
        )

    def forward(self, x):
        ff_out = self.adapter_block(x)
        adapter_out = ff_out + x
        return adapter_out

class Adaptered(nn.Module):
    def __init__(self, orig_layer, domains=[], domain_controller=None, alternative=0, no_base=False):
        super().__init__()
        self.orig_layer = orig_layer
        self.adapter = Adapter()
        self.domains = domains
        self.domain_controller = domain_controller
        self.no_base = no_base
        # if not self.no_base:
        self.domain_adapters = nn.ModuleDict({domain: DomainAdapter() for domain in self.domains})
        self.alternative = alternative

    def average_adapter(self, adapters):
      total_adapters = len(adapters)
      weights1 = []
      weights2 = []

      # Average the weights of the adapters.
      for adapter in adapters:
          weights1.append(adapter.adapter_block[0].weight)
          weights2.append(adapter.adapter_block[2].weight)

      averaged_weights1 = torch.mean(torch.stack(weights1), dim=0)
      averaged_weights2 = torch.mean(torch.stack(weights2), dim=0)

      # Get the weights of the two adapters.
      new_linear1 = nn.Linear(768, 128)
      new_linear1.weight = nn.Parameter(averaged_weights1)
      new_linear2 = nn.Linear(128, 768)
      new_linear2.weight = nn.Parameter(averaged_weights2)


      # Create a new adapter with the averaged weights.
      new_adapter = nn.Sequential(
        new_linear1,
        nn.ReLU(),
        new_linear2,
      )
      new_adapter.to("cuda")

      return new_adapter

    def forward(self, *x):
        orig_out = self.orig_layer(*x)
        if self.no_base:
          return orig_out
        else:
          base_output = self.adapter.forward(orig_out)
          if self.domain_controller.current_batch_domain not in self.domains:
            if self.alternative == 0:
              return base_output
            elif self.alternative == 1:
              domain_adapters_list = [module for name, module in self.domain_adapters.items()]
              average_domain_adapter = self.average_adapter(domain_adapters_list)
              return average_domain_adapter(orig_out)
            elif self.alternative == 2:
              domain_adapters_list = [module for name, module in self.domain_adapters.items()]
              average_domain_adapter = self.average_adapter(domain_adapters_list)
              averaged_weights1 = torch.mean(torch.stack([self.adapter.adapter_block[0].weight, average_domain_adapter[0].weight]), dim=0)
              averaged_weights2 = torch.mean(torch.stack([self.adapter.adapter_block[2].weight, average_domain_adapter[2].weight]), dim=0)
              new_linear1 = nn.Linear(768, 128)
              new_linear1.weight = nn.Parameter(averaged_weights1)
              new_linear2 = nn.Linear(128, 768)
              new_linear2.weight = nn.Parameter(averaged_weights2)
              average_base_domain_adapter = nn.Sequential(
                new_linear1,
                nn.ReLU(),
                new_linear2,
              )
              average_base_domain_adapter.to("cuda")
              return average_base_domain_adapter(orig_out)
          else:
              current_domain_adapter = self.domain_adapters[self.domain_controller.current_batch_domain]
              domain_output = current_domain_adapter.forward(base_output)
              return domain_output


class GPT2_with_adapter(nn.Module):
    def __init__(self, domains=[], single_domain=None, no_adapter=False, alternative=0):
        super().__init__()
        self.model = AutoModelForCausalLM.from_pretrained("gpt2")
        self.domain_controller = DomainController()
        self.domains = domains
        self.no_adapter = no_adapter
        self.single_domain = single_domain
        self.alternative = alternative
        if single_domain:
          self.domain_controller.current_batch_domain = single_domain

        # Embed adapter layers into the transformer blocks
        if not self.no_adapter:
          # Freeze the original model parameters
          for params in self.model.parameters():
            params.requires_grad = False
          for i in range(12):
              self.model.transformer.h[i].mlp.c_proj = Adaptered(self.model.transformer.h[i].mlp.c_proj, domains, self.domain_controller, self.alternative)

    def get_model(self):
        return self.model

    def forward(self, input_ids, attention_mask, labels=None, batch_domain=None):
        if batch_domain and not self.single_domain and not self.no_adapter:
          # assert batch_domain in self.domains
          self.domain_controller.current_batch_domain = batch_domain
        return self.model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
```

```python
from transformers import Trainer, TrainingArguments
from torch.utils.data.distributed import DistributedSampler
from torch.utils.data.sampler import RandomSampler
from typing import List, Union, Dict
import numpy as np

class DataLoaderWithTaskname:
    """
    Wrapper around a DataLoader to also yield a task name
    """
    def __init__(self, batch_domain, data_loader):
        self.batch_domain = batch_domain
        self.data_loader = data_loader

        self.batch_size = data_loader.batch_size
        self.dataset = data_loader.dataset

    def __len__(self):
        return len(self.data_loader)

    def __iter__(self):
        for batch in self.data_loader:
            batch["batch_domain"] = self.batch_domain
            yield batch


class MultitaskDataloader:
    """
    Data loader that combines and samples from multiple single-task
    data loaders.
    """
    def __init__(self, dataloader_dict):
        self.dataloader_dict = dataloader_dict
        self.num_batches_dict = {
            batch_domain: len(dataloader)
            for batch_domain, dataloader in self.dataloader_dict.items()
        }
        self.domain_name_list = list(self.dataloader_dict)
        self.dataset = [None] * sum(
            len(dataloader.dataset)
            for dataloader in self.dataloader_dict.values()
        )

    def __len__(self):
        return sum(self.num_batches_dict.values())

    def __iter__(self):
        """
        For each batch, sample a task, and yield a batch from the respective
        task Dataloader.

        We use size-proportional sampling, but you could easily modify this
        to sample from some-other distribution.
        """
        domain_choice_list = []
        for i, domain_name in enumerate(self.domain_name_list):
            domain_choice_list += [i] * self.num_batches_dict[domain_name]
        domain_choice_list = np.array(domain_choice_list)
        np.random.shuffle(domain_choice_list)
        dataloader_iter_dict = {
            domain_name: iter(dataloader)
            for domain_name, dataloader in self.dataloader_dict.items()
        }
        for domain_choice in domain_choice_list:
            domain_name = self.domain_name_list[domain_choice]
            yield next(dataloader_iter_dict[domain_name])

class MultitaskTrainer(Trainer):

    def get_single_train_dataloader(self, domain_name, train_dataset):
        """
        Create a single-task data loader that also yields task names
        """
        if self.train_dataset is None:
            raise ValueError("Trainer: training requires a train_dataset.")
        else:
            train_sampler = (
                RandomSampler(train_dataset)
                if self.args.local_rank == -1
                else DistributedSampler(train_dataset)
            )

        data_loader = DataLoaderWithTaskname(
            batch_domain=domain_name,
            data_loader=DataLoader(
              train_dataset,
              batch_size=self.args.train_batch_size,
              sampler=train_sampler,
              collate_fn=self.data_collator,
            ),
        )

        return data_loader

    def get_train_dataloader(self):
        """
        Returns a MultitaskDataloader, which is not actually a Dataloader
        but an iterable that returns a generator that samples from each
        task Dataloader
        """
        return MultitaskDataloader({
            task_name: self.get_single_train_dataloader(task_name, task_dataset)
            for task_name, task_dataset in self.train_dataset.items()
        })
```

```python
zero_shot_model = AutoModelForCausalLM.from_pretrained("gpt2")
zero_shot_model.eval()
```

```python
import math
import collections
from tqdm import tqdm
training_args = TrainingArguments(
    output_dir="/content/gdrive/MyDrive/s2orc_results_june6/zero_shot",
    evaluation_strategy="epoch",
    learning_rate=1e-4,
    num_train_epochs=1,
    weight_decay=0.01,
)

test_datasets = {"psych": lm_psych_abstract,
                  "cs": lm_cs_abstract,
                  "math": lm_math_abstract,
                  "med": lm_med_abstract,
                  "geo": lm_geo_abstract,
                  "art": lm_art_abstract,
                  "business": lm_business_abstract,
                  "history": lm_history_abstract
                 }

zero_shot_results = collections.defaultdict(list)

for domain, dataset in tqdm(test_datasets.items()):
    trainer = Trainer(
      model=zero_shot_model,
      args=training_args,
      train_dataset=dataset['train'],
      eval_dataset=dataset['test'],
      data_collator=data_collator,
    )

    zero_shot_results[domain].append(math.exp(trainer.evaluate()['eval_loss']))
```
